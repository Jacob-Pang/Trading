# 1. Geometric Brownian Motion (GBM)

Note that $\mu$ and $\sigma$ are based on logarithmic returns<br>
For&ensp;$t_{1} \leq t_{2}$&ensp;and&ensp;$\delta t := t_{1} \rightarrow t_{2}, \; |\delta t| = t_{2} - t_{1}$

$$
\ln{S_{t_{2}}} = \ln{S_{t_{1}}} +
\left (
    \mu_{\mathrm{S_{\delta t}}} -
    \frac{\sigma_{\mathrm{S_{\delta t}}} ^ 2}{2}
\right ) \delta t +
\sigma_{\mathrm{S_{\delta t}}} \, W_{\delta t}
$$

<br>

$$
\ln{S_{t_{2}} | S_{t_{1}}} \sim \mathbb{N} \left(
    \ln{S_{t_{1}}} +
    \left (
        \mu_{\mathrm{S_{\delta t}}} -
        \frac{\sigma_{\mathrm{S_{\delta t}}} ^ 2}{2}
    \right ) \delta t, \;
    \sigma_{\mathrm{S_{\delta t}}}^2 \delta t
\right )
$$

$$
\ln{\frac{S_{t_{2}} | S_{t_{1}}}{{S_{t_{1}}}}} \sim \mathbb{N} \left(
    \left (
        \mu_{\mathrm{S_{\delta t}}} -
        \frac{\sigma_{\mathrm{S_{\delta t}}} ^ 2}{2}
    \right ) \delta t, \;
    \sigma_{\mathrm{S_{\delta t}}}^2 \delta t
\right )
$$

For covenience,&ensp;$\ln{S_{t} | S_{0}} := \ln{S_{t}}$

$$
\ln{S_{t}} \sim \mathbb{N} \left(
    \ln{S_{0}} +
    \left (
        \mu_{\mathrm{S_{t}}} -
        \frac{\sigma_{\mathrm{S_{t}}} ^ 2}{2}
    \right ) t, \;
    \sigma_{\mathrm{S_{t}}}^2 t
\right )
$$

$$
\ln{\frac{S_{t}}{S_{0}}}  \sim \mathbb{N} \left(
    \left (
        \mu_{\mathrm{S_{t}}} -
        \frac{\sigma_{\mathrm{S_{t}}} ^ 2}{2}
    \right ) t, \;
    \sigma_{\mathrm{S_{t}}}^2 t
\right )
$$

Note that&ensp;$S_{t_{1}}$&ensp;and the increment&ensp;
$
\begin{aligned}
\frac{S_{t_{2}}}{S_{t_{1}}}
\end{aligned}
$
&ensp;are independent

Then,&ensp;
$
\begin{aligned}
\ln{S_{t_{2}}} = \ln{S_{t_{1}}} + \ln{\frac{S_{t_{2}}}{S_{t_{1}}}}
\end{aligned}
$
&ensp;is a linear combination of independent gaussian RVs

$$
\ln{S_{t_{2}}} \sim \mathbb{N} \left(
    \ln{S_{0}} +
    \left (
        \mu_{\mathrm{S_{t_{1}}}} -
        \frac{\sigma_{\mathrm{S_{t_{1}}}} ^ 2}{2}
    \right ) t_{1} +
    \left (
        \mu_{\mathrm{S_{\delta t}}} -
        \frac{\sigma_{\mathrm{S_{\delta t}}} ^ 2}{2}
    \right ) \delta t, \;
    \sigma_{\mathrm{S_{t}}}^2 t_{1} +
    \sigma_{\mathrm{S_{\delta t}}}^2 \delta t
\right )
$$

## 1.1. Risk Neutral Measure

Sets&ensp;$\mu_{\mathrm{S_{\delta t}}} = r$,&ensp;the risk-free rate

$$
\ln{Q_{t_{2}} | S_{t_{1}}} \sim \mathbb{N} \left(
    \ln{S_{t_{1}}} +
    \left (
        r - \frac{\sigma_{\mathrm{S_{\delta t}}} ^ 2}{2}
    \right ) \delta t, \;
    \sigma_{\mathrm{S_{\delta t}}}^2 \delta t
\right )
$$

## 1.2. Expectation and Variance

$$
\mathbb{E} \left [ S_{t} \right ] = e^{\left(
    \ln{S_{0}} + \mu_{\mathrm{S_{t}}} + \frac{\sigma_{\mathrm{S_{t}}}^2}{2}
\right )} = S_{0} \, e ^ {\mu_{\mathrm{S_{t}}} t}
$$

$$\mathbb{V} \left [ S_{t} \right ] = S_{0}^{2} \, e ^ {2 \mu_{\mathrm{S_{t}}} t} \left ( e ^ {\sigma^2 t} - 1\right )$$

## 1.3. Covariance

$
\begin{aligned}
\mathrm{Cov} \left (S_{t}^a, \, S_{t}^b \right ) =
\mathbb{E} \left [ S_{t}^a \, S_{t}^b \right ] -
\mathbb{E} \left [ S_{t}^a \right ] \, \mathbb{E} \left [ S_{t}^b \right ]
\end{aligned}
$

> <br>Let $U = S_{t}^{a} \, S_{t}^{b}$
> <br><br>
$
\begin{aligned}
\Rightarrow \ln{U} = \ln{S_{t}^{a}} + \ln{S_{t}^{b}}
\end{aligned}
$
> <br> <br> Note that&ensp;
$
\begin{aligned}
\ln{S_{t}^{a}} \sim \mathbb{N} \left(
    \ln{S_{0}^a} + \mu_{\mathrm{S_{t}^{a}}}, \,
    \sigma_{\mathrm{S_{t}^{a}}}^2
\right) ,
\end{aligned}
$
&ensp;
$
\begin{aligned}
\ln{S_{t}^{b}} \sim \mathbb{N} \left(
    \ln{S_{0}^b} + \mu_{\mathrm{S_{t}^{b}}}, \,
    \sigma_{\mathrm{S_{t}^{b}}}^2
\right)
\end{aligned}
$
> <br> <br>
$
\begin{aligned}
\Rightarrow \ln{U} = \ln{S_{t}^{a}} + \ln{S_{t}^{b}} \sim \mathbb{N} \left (
    \ln{S_{0}^a} + \ln{S_{0}^b} + \mu_{\mathrm{S_{t}^{a}}} + \mu_{\mathrm{S_{t}^{b}}}, \,
    \sigma_{\mathrm{S_{t}^{a}}}^2 +
    \sigma_{\mathrm{S_{t}^{b}}}^2 +
    2 \, \rho_{\left ( \ln{S_{t}^a}, \, \ln{S_{t}^b} \right )} \,
    \sigma_{\mathrm{S_{t}^{a}}} \, \sigma_{\mathrm{S_{t}^{b}}}
\right )
\end{aligned}
$
> <br> <br>
$
\begin{aligned}
\Rightarrow \mathbb{E} \left [ {U} \right ] = \mathrm{exp} \left (
    \ln{S_{0}^a} + \ln{S_{0}^b} + \mu_{\mathrm{S_{t}^{a}}} + \mu_{\mathrm{S_{t}^{b}}} + \frac{
        \sigma_{\mathrm{S_{t}^{a}}}^2 + \sigma_{\mathrm{S_{t}^{b}}}^2 +
        2 \, \rho_{\left ( \ln{S_{t}^a}, \, \ln{S_{t}^b} \right )} \, \sigma_{\mathrm{S_{t}^{a}}} \, \sigma_{\mathrm{S_{t}^{b}}}
    }{2}
\right )
\end{aligned}
$
> <br> <br> &emsp; &emsp;
$
\begin{aligned}
= \mathrm{exp} \left (
    \left [ \ln{S_{0}^a} + \mu_{\mathrm{S_{t}^{a}}} + \frac{\sigma_{\mathrm{S_{t}^{a}}}^2}{2} \right ]+
    \left [ \ln{S_{0}^b} + \mu_{\mathrm{S_{t}^{b}}} + \frac{\sigma_{\mathrm{S_{t}^{b}}}^2}{2} \right ]+
    \rho_{\left ( \ln{S_{t}^a}, \, \ln{S_{t}^b} \right )} \, \sigma_{\mathrm{S_{t}^{a}}} \, \sigma_{\mathrm{S_{t}^{b}}}
\right )
\end{aligned}
$
> <br> <br> &emsp; &emsp;
$
\begin{aligned}
= \mathbb{E} \left [ S_{t}^a \right ] \,
\mathbb{E} \left [ S_{t}^b \right ] \,
\mathrm{exp} \left ( \mathrm{Cov} \left ( \ln{S_{t}^a}, \, \ln{S_{t}^b} \right ) \right )
\end{aligned}
$
> <br> <br> &emsp; &emsp;
$
\begin{aligned}
= S_{0}^a \, S_{0}^b \, \mathrm{exp} \left (
    \left ( \mu_{\mathrm{S_{t}^a}} + \mu_{\mathrm{S_{t}^b}} \right ) \, t \, + \,
    \mathrm{Cov} \left ( \ln{S_{t}^a}, \, \ln{S_{t}^b} \right )
\right )
\end{aligned}
$
> <br><br>

$
\begin{aligned}
\Rightarrow \mathrm{Cov} \left (S_{t}^a, \, S_{t}^b \right ) =
\mathbb{E} \left [ S_{t}^a \right ] \, \mathbb{E} \left [ S_{t}^b \right ] \, \left [
    \mathrm{exp} \left ( \mathrm{Cov} \left ( \ln{S_{t}^a}, \, \ln{S_{t}^b} \right ) \right ) - 1
\right ]
\end{aligned}
$

&emsp; &emsp;
$
\begin{aligned}
= S_{0}^a \, S_{0}^b \,
\mathrm{exp} \left ( \left (
    \mu_{\mathrm{S_{t}^a}}
    + \mu_{\mathrm{S_{t}^b}}
\right ) \, t
\right ) \,
\left [
    \mathrm{exp} \left ( \mathrm{Cov} \left ( \ln{S_{t}^a}, \, \ln{S_{t}^b} \right ) \right ) - 1
\right ]
\end{aligned}
$

### 1.3.1 Covariance of $\ln{S_{t}}$ from Non-Logarithmic Returns

$
\begin{aligned}
\mathrm{Cov} \left (
    \frac{S_{t}^a}{S_{0}^a} - 1, \,
    \frac{S_{t}^b}{S_{0}^b} - 1
\right ) = \mathrm{Cov} \left (
    \frac{S_{t}^a}{S_{0}^a}, \,
    \frac{S_{t}^b}{S_{0}^b}
\right )
\end{aligned}
$

&emsp; &emsp;
$
\begin{aligned}
= \frac{ \mathrm{Cov} \left ( S_{t}^a, \, S_{t}^b \right )}{S_{0}^a \, S_{0}^b}
\end{aligned}
$

&emsp; &emsp;
$
\begin{aligned}
= e ^ {\left(\mu_{a} + \mu_{b} \right ) t} \, \left [
    \mathrm{exp} \left ( \mathrm{Cov} \left ( \ln{S_{t}^a}, \, \ln{S_{t}^b} \right ) \right ) - 1
\right ]
\end{aligned}
$

&emsp; &emsp;
$
\begin{aligned}
\mathrm{Cov} \left (
    \ln{\frac{S_{t}^a}{S_{0}^a}}, \,
    \ln{\frac{S_{t}^b}{S_{0}^b}}
\right ) = \mathrm{Cov} \left (\ln{S_{t}^a}, \, \ln{S_{t}^b} \right )
\end{aligned}
$

&emsp; &emsp; &emsp; &emsp;
$
\begin{aligned}
= \ln{\left [
    \frac{
        \mathrm{Cov} \left (
            \frac{S_{t}^a}{S_{0}^a}, \,
            \frac{S_{t}^b}{S_{0}^b}
        \right )
    }{e ^ {\left(
        \mu_{\mathrm{S_{t}^a}} +
        \mu_{\mathrm{S_{t}^b}}
    \right ) t}} + 1
\right ]}
\end{aligned}
$

&emsp; &emsp;
Note that $\mathrm{Cov} \left (\ln{S_{t}^a}, \, \ln{S_{t}^b} \right ) = t \, \mathrm{Cov} \left (\ln{S_{1}^a}, \, \ln{S_{1}^b} \right )$

# 2. Options

For an option with time to expiry&ensp;$T \geq t$&ensp;and strike price&ensp;$K$ ,&ensp;the price at time&ensp;$t$&ensp;is the present value of the expectation of the payoff at time&ensp;$T$&ensp;based on risk-neutral measure conditioned on the current spot price,&ensp;$S_{t}$&ensp;and assumes volatility to be constant&ensp;$\sigma_{\mathrm{S_{t}}} = \sigma_{\mathrm{S_{\delta t}}}$

$$
C_{t} = 
e^{-r \delta t} \,
\mathbb{E} \left [
    \mathrm{max} \left \{ Q_{T} - K, \, 0 \right \} | \, S_{t}
\right ]
$$

$$
P_{t} = 
e^{-r \delta t} \,
\mathbb{E} \left [
    \mathrm{max} \left \{ K - Q_{T} , \, 0 \right \} | \, S_{t}
\right ]
$$

$$
C_{t} + K e^{-r\delta t} = P_{t} + S_{t}
$$

<br>

$$
\ln{Q_{T} | S_{t}} \sim \mathbb{N} \left(
    \ln{S_{t}} +
    \left (
        r - \frac{\sigma_{\mathrm{S_{t}}} ^ 2}{2}
    \right ) \delta t, \;
    \sigma_{\mathrm{S_{t}}}^2 \delta t
\right )
$$

$$
\ln{Q_{T}} \sim \mathbb{N} \left(
    \ln{S_{0}} +
    \left (
        \mu_{\mathrm{S_{t}}} -
        \frac{\sigma_{\mathrm{S_{t}}} ^ 2}{2}
    \right ) t +
    \left (
        r - \frac{\sigma_{\mathrm{S_{t}}} ^ 2}{2}
    \right ) \delta t, \;
    \sigma_{\mathrm{S_{t}}}^2 \, T
\right )
$$

$
\begin{aligned}
C_{t} = 
e^{-r \delta t} \,
\mathbb{E} \left [
    \mathrm{max} \left \{ Q_{T} - K, \, 0 \right \} | \, S_{t}
\right ]
\end{aligned}
$

&emsp; &emsp;
$
\begin{aligned}
= e^{-r \delta t} \,
\int_{-\infty}^{+\infty}
    \mathrm{max} \left \{ x - K, \, 0 \right \} \,
    f_{Q_{T} | \, S_{t}} \left ( x \right )
dx
\end{aligned}
$

&emsp; &emsp;
$
\begin{aligned}
= e^{-r \delta t} \, \left [
    \int_{K}^{+\infty}
        x \,
        f_{Q_{T} | \, S_{t}} \left ( x \right )
    dx - K \, \int_{K}^{+\infty}
        f_{Q_{T} | \, S_{t}} \left ( x \right )
    dx
\right ]
\end{aligned}
$

&emsp; &emsp; Sub (LN1.1)&ensp;
$
\begin{aligned}
\mu = \ln{S_{t}} +
    \left (
        r - \frac{\sigma_{\mathrm{S_{t}}} ^ 2}{2}
    \right ) \delta t \; , \;
\sigma = \sigma_{\mathrm{S_{t}}} \sqrt{\delta t} \; , \;
k_{1} = 1 \; , \;
k_{2} = 0
\end{aligned}
$

&emsp; &emsp;
$
\begin{aligned}
= \mathrm{exp} \left(
    \ln{S_{t}} +
    \left (
        r - \frac{\sigma_{\mathrm{S_{t}}} ^ 2}{2}
    \right ) \delta t +
    \frac {\sigma_{\mathrm{S_{t}}}^2}{2} \delta t -
    r \delta t
\right ) \left [
    1 - \Phi \left (
        \frac{\ln{K} - \left [
            \ln{S_{t}} +
            \left (
                r - \frac{\sigma_{\mathrm{S_{t}}} ^ 2}{2}
            \right ) \delta t +
            \sigma_{\mathrm{S_{t}}}^2 \delta t
        \right ]}
        {\sigma_{\mathrm{S_{t}}} \sqrt{\delta t}}
    \right )
\right ] \, -
\end{aligned}
$

&emsp; &emsp; &emsp; &emsp;
$
\begin{aligned}
K \, e^{-r \delta t} \, \left [
    1 - \Phi \left (
        \frac{
            \ln{K} -
            \ln{S_{t}} -
            \left (
                r - \frac{\sigma_{\mathrm{S_{t}}} ^ 2}{2}
            \right ) \delta t
        }{\sigma_{\mathrm{S_{t}}} \sqrt{\delta t}}
    \right )
\right ]
\end{aligned}
$

&emsp; &emsp;
$
\begin{aligned}
= S_{t} \, \Phi \left ( d_{1, t} \right ) -
K \, e^{-r \delta t} \, \Phi \left ( d_{2, t} \right )
\end{aligned}
$

$$
d_{1, t} := \frac{
    \ln{S_{t}} -
    \ln{K} + \left(
        r + \frac{\sigma_{\mathrm{S_{t}}}^2}{2}
    \right ) \delta t
}{
    \sigma_{\mathrm{S_{t}}} \,
    \sqrt{\delta t}
}
$$

$$
d_{2, t} := d_{1, t} - \sigma_{\mathrm{S_{t}}} \, \sqrt{\delta t}
$$

&emsp; &emsp; Note that the formula decays to the Black Scholes formula when&ensp;$t = 0 \Rightarrow |\delta t| = T$

Applying Put-Call Parity

$
\begin{aligned}
P_{t} = C_{t} + K e^{-r\delta t} - S_{t}
\end{aligned}
$

&emsp; &emsp;
$
\begin{aligned}

\end{aligned}
$

&emsp; &emsp;
$
\begin{aligned}

\left(
    \Phi \left ( d_{1, t} \right ) - 1

\left (
    \Phi \left ( d_{2, t} \right ) - 1
\right )
\end{aligned}
$

&emsp; &emsp;
$
\begin{aligned}

\Phi \left ( - d_{1, t} \right )
\end{aligned}
$

## 2.3. Covariance

Let&ensp;
$
\begin{aligned}
\vec{S_{t}} := \left < S_{t}^{a} \, S_{t}^b \right>
\end{aligned}
$

&emsp; &emsp;
$
\begin{aligned}
\mu_{a} := \ln{S_{0}^a} + \left (
    \mu_{\mathrm{S_{t}^a}} -
    \frac{\sigma_{\mathrm{S_{t}^a}} ^ 2}{2}
\right ) t, \;
\rho := \rho_{\ln{S_{t}^a}, \ln{S_{t}^b}}
\end{aligned}
$

&emsp; &emsp;
$
\begin{aligned}
\beta_{1}^a := \ln{K_{a}} - \left(
    r_{a} + \frac{\sigma_{\mathrm{S_{t}^a}}^2}{2}
\right ) \delta t_{a}, \;
\beta_{2}^a := \beta_{1}^a + \sigma_{\mathrm{S_{t}^a}}^2 \delta t_{a}
\end{aligned}
$

Note that for the case&ensp;$\rho = 0 \; | \; \sigma_{\mathrm{S_{t}^x}} = 0 \; | \; t = 0$,&ensp;covariance&ensp;$= 0$.

Henceforth, we only consider the case where

$$\rho \neq 0 \; , \; \sigma_{\mathrm{S_{t}^x}} \, , \, t > 0$$

### 2.3.1. Covariance of Stock and Option

$
\begin{aligned}
\mathrm{Cov} \left( S_{t}^{a}, \, C_{t}^{b} \right ) =
\mathbb{E} \left [ S_{t}^{a} \, C_{t}^{b} \right ] -
\mathbb{E} \left [ S_{t}^{a} \right ] \,
\mathbb{E} \left [ C_{t}^{b} \right ]
\end{aligned}
$

&emsp; &emsp;
$
\begin{aligned}
= \mathbb{E} \left [
    S_{t}^{a} \, S_{t}^b \,
    \Phi \left (
        \frac{
            \ln{S_{t}^b} - \beta_{1}^b
        }{\sigma_{\mathrm{S_{t}^b}} \sqrt{\delta t_{b}}}
    \right )
\right ] - K_{b} \, e^{-r_{b} \, \delta t_{b}} \,
\mathbb{E} \left [
    S_{t}^{a} \,
    \Phi \left (
        \frac{
            \ln{S_{t}^b} - \beta_{2}^b
        }{\sigma_{\mathrm{S_{t}^b}} \sqrt{\delta t_{b}}}
    \right )
\right ] \, -
\mathbb{E} \left [ S_{t}^{a} \right ] \,
\mathbb{E} \left [ C_{t}^{b} \right ]
\end{aligned}
$

&emsp; &emsp;
$
\begin{aligned}
= S_{0}^a \, S_{0}^b \,
\mathrm{exp} \left (
    \mu_{ \mathrm{S_{t}^a}} \, t +
    \mu_{\mathrm{S_{t}^b}} \, t +
    \mathrm{Cov} \left(
        \ln{S_{t}^a}, \, \ln{S_{t}^b}
    \right )
\right ) \,
\Phi \left (
    \frac{
        \ln{S_{0}^b} -
        \ln{K_{b}} +
        \mu_{\mathrm{S_{t}^b}} \, t +
        r_{b} \, \delta t_{b} +
        \mathrm{Cov} \left(
            \ln{S_{t}^a}, \, \ln{S_{t}^b}
        \right ) +
        \frac{\sigma_{\mathrm{S_{t}^b}}^2}{2} \, T_{b}
    }
    {
        \sigma_{\mathrm{S_{t}^b}} \,
        \sqrt{T_{b}}
    }
\right ) \, -
\end{aligned}
$

&emsp; &emsp; &emsp; &emsp;
$
\begin{aligned}
K_{b} \, S_{0}^a \, e^ {
    \left(
        \mu_{\mathrm{S_{t}^a}} \, t - r_{b} \, \delta t_{b}
    \right )
} \,
\Phi \left (
    \frac{
        \ln{S_{0}^b} -
        \ln{K_{b}} +
